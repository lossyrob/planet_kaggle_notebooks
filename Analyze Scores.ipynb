```python
def scores_to_df(raw_scores):
    cols = ['name', 'f2_samples', 'f2_labels', 'lb_score',
            'clear', 'cloudy', 'haze', 'partly_cloudy',
            'agriculture', 'bare_ground', 'cultivation', 'habitation', 'primary', 'road', 'water',
            'artisinal_mine', 'blooming', 'blow_down', 'conventional_mine', 'selective_logging', 'slash_burn']
    df = None
    for name in raw_scores:
        s = raw_scores[name]
        scores = {}
        scores['name'] = name
        if "lb_score" in s:
            scores['lb_score'] = s['lb_score']
        else:
            scores['lb_score'] = None
            
        if "atmos_scores" in s:
            scores['f2_samples'] = s['f2_samples']
            scores['f2_labels'] = s['f2_labels']
            for k in s['atmos_scores']:
                scores[k] = s['atmos_scores'][k]

            for k in s['common_scores']:
                scores[k] = s['common_scores'][k]

            for k in s['rare_scores']:
                scores[k] = s['rare_scores'][k]

        else:
            scores['f2_samples'] = s['f2']
            scores['f2_labels'] = None

            for k in s['f2_subscores']:
                scores[k] = s['f2_subscores'][k]

        this_df = pd.DataFrame([scores], columns=list(scores.keys()))
        if df is None: 
            df = this_df
        else:
            df = df.append(this_df, ignore_index=True)
        df = df[cols]
    return df
```

```python
# Setting the positions and width for the bars
pos = list(range(len(df['name'])))
width = 0.25

# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos,
        #using df['pre_score'] data,
        df['f2_samples'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#EE3224',
        # with label the first value in first_name
        label=df['name'][0])

plt.bar(pos,
        #using df['pre_score'] data,
        df['clear'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F78F1E',
        # with label the first value in first_name
        label=df['name'][1])

# Set the y axis label
ax.set_ylabel('Score')

# Set the chart's title
ax.set_title('Model Scores')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['name'],rotation='vertical')

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
def calc(labels, func):
    df2 = df[labels[0]].copy()
    for l in labels[1:]:
        df2.append(df[l])
    return func(df2)
plt.ylim([calc(['f2_samples', 'clear'], min) - 0.03, calc(['f2_samples', 'clear'], max) + 0.01] )

# Adding the legend and showing the plot
plt.legend(['Pre Score', 'Mid Score', 'Post Score'], loc='upper left')
plt.grid()
plt.show()
```
